```python
from sklearn.datasets import load_digits
Data=load_digits()
Features=Data.data
Targets=Data.target
```

```python
from sklearn.model_selection import train_test_split
X_Train,X_Test,Y_Train,Y_Test=train_test_split(Features,Targets,test_size=0.2,random_state=0)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,classification_report
import matplotlib.pyplot as plt
import seaborn as sns
```

### Using Linear Kernel

```python
Lin=SVC(kernel='linear')
Lin.fit(X_Train,Y_Train)
pred_Lin=Lin.predict(X_Test)
Mat=confusion_matrix(pred_Lin,Y_Test)
sns.heatmap(Mat,annot=True)
plt.show()
print(classification_report(pred_Lin,Y_Test))
```

### Using Poly kernel

```python
quad=SVC(kernel='poly')
quad.fit(X_Train,Y_Train)
pred_quad=quad.predict(X_Test)
Mat=confusion_matrix(pred_quad,Y_Test)
sns.heatmap(Mat,annot=True)
plt.show()
print(classification_report(pred_quad,Y_Test))
```

### Using Gaussian Kernel

```python
rbf=SVC()
rbf.fit(X_Train,Y_Train)
pred_rbf=rbf.predict(X_Test)
Mat=confusion_matrix(pred_rbf,Y_Test)
sns.heatmap(Mat,annot=True)
plt.show()
print(classification_report(pred_rbf,Y_Test))
```

### Using Sigmoid Kernel

```python
sig=SVC(kernel='sigmoid')
sig.fit(X_Train,Y_Train)
pred_sig=sig.predict(X_Test)
Mat=confusion_matrix(pred_sig,Y_Test)
sns.heatmap(Mat,annot=True)
plt.show()
print(classification_report(pred_sig,Y_Test))
```
